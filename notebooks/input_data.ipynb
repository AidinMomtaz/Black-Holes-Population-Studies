```python
import pandas as pd
from pathlib import Path

data_folder = "D:\iorio+22_Zenodo_repository_V2\iorio+22_Zenodo_repository_V2\data_from_simulations\simulation_F"

df = pd.read_csv(Path(data_folder,'summary.csv'))

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(figsize=(15,10), nrows=2)


sns.lineplot(data = df, x = 'Z', y='BHBHm',ax = ax[0],
            hue= 'alpha',
            palette="ch:rot=-.25,hue=1,light=.75")

sns.lineplot(data = df, x = 'Z', y='BHBH',ax = ax[1],
            hue= 'alpha',
            palette="ch:rot=-.25,hue=1,light=.75")
```

```python
fig, ax = plt.subplots(figsize=(15,10))

sns.scatterplot(data=df,  x='Z', y='alpha', size="BHBHm", ax=ax,sizes=(400, 10))
ax.set_xscale('log')
ax.set_yscale('log')
```
